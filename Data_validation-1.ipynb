```python
import pandas as pd
import random
import requests
from bs4 import BeautifulSoup
from googlesearch import search
import urllib.parse
```

```python
df = pd.read_excel("/content/ALY6080_DataValidation_Group_4.xlsx")
```

```python
import requests
import urllib.parse

# Iterate over each row in the dataset
for index, row in df.iterrows():
    parent_company_name = str(row['parent_company_name'])

    # Clean parent company name
    parent_company_name = urllib.parse.quote_plus(parent_company_name)

    # Search for the parent company name on Wikidata
    search_url = f"https://www.wikidata.org/w/api.php?action=wbsearchentities&search={parent_company_name}&format=json"
    response = requests.get(search_url).json()

    # Check if search results are found
    if 'search' in response and len(response['search']) > 0:
        first_result = response['search'][0]
        page_id = first_result['id']
        page_url = f"https://www.wikidata.org/wiki/{page_id}"

        # Check if the page exists
        page_exists_url = f"https://www.wikidata.org/w/api.php?action=query&format=json&titles={page_id}"
        page_exists_response = requests.get(page_exists_url).json()
        page_exists = 'missing' not in page_exists_response['query']['pages']

        # Generate the URL in the desired format
        url = f"https://www.wikidata.org/wiki/{page_id}"

        # Update the row in the dataset
        df.at[index, 'id'] = page_id
        df.at[index, 'link'] = url if page_exists else ''
        df.at[index, 'validated'] = page_exists
    else:
        # If no search results are found, set 'validated' to False
        df.at[index, 'validated'] = False

# Display the updated dataset
print(df[['parent_company_name', 'id', 'link', 'country', 'sum']].to_string(index=False))
```

```python
import requests
import urllib.parse

# Function to check if a page exists on Wikidata
def check_page_exists(page_id):
    page_exists_url = f"https://www.wikidata.org/w/api.php?action=query&format=json&titles={page_id}"
    page_exists_response = requests.get(page_exists_url).json()
    page_exists = 'missing' not in page_exists_response['query']['pages']
    return page_exists

# Iterate over each row in the dataset
for index, row in df.iterrows():
    parent_company_name = str(row['parent_company_name'])

    # Clean parent company name
    parent_company_name = urllib.parse.quote_plus(parent_company_name)

    # Search for the parent company name on Wikidata
    search_url = f"https://www.wikidata.org/w/api.php?action=wbsearchentities&search={parent_company_name}&format=json"
    response = requests.get(search_url).json()

    # Check if search results are found
    if 'search' in response and len(response['search']) > 0:
        first_result = response['search'][0]
        page_id = first_result['id']

        # Check if the page ID is valid
        if page_id.startswith('Q'):
            # Generate the shortened URL
            shortened_url = f"https://www.wikidata.org/wiki/{page_id}"

            # Check if the page exists
            page_exists = check_page_exists(page_id)

            # Update the row in the dataset
            df.at[index, 'id'] = page_id
            df.at[index, 'link'] = shortened_url if page_exists else ''
            df.at[index, 'validated'] = page_exists
        else:
            # If the page ID is not valid, set 'validated' to False
            df.at[index, 'validated'] = False
    else:
        # If no search results are found, set 'validated' to False
        df.at[index, 'validated'] = False

# Display the updated dataset
print(df[['parent_company_name', 'id', 'link', 'country', 'sum']].to_string(index=False))
```
